We've been exploring our data in a 2D format but we need to reshape our data back into a 3D numpy array for tensorflow. The array's shape is going to end up being 1260 (total number of video samples) x 40 (total frames per video) x 1662 (total landmarks per frame).

The below function does this for us.

```python
def mat_to_3d(some_df):
    num_of_cols = some_df.shape[1]
    len_of_df = some_df.shape[0]
    my_array = np.zeros((1260, 40, num_of_cols))
    
    for i, j in enumerate(np.arange(0, len_of_df, 40)):
        subset_array = np.arange(j, j+40)
        new_df = some_df.iloc[subset_array]
        my_array[i] = new_df.to_numpy()

    return my_array
```

```python
# convert back to 3d array
all_lm = mat_to_3d(all_lm)
# sanity check
all_lm.shape
```

#### Modelling

The first modelling we'll be doing is on all of the landmark location data. The first LSTM model will be trained on 1000 epochs.

Let's first create our training, validation, and test sets.

```python
# Need to change labels_numerical to shape 1260x6
target_labels = to_categorical(labels_numerical).astype(int)

X_resid_train_1, X_test_1, y_resid_train_1, y_test_1 = train_test_split(all_lm, target_labels, random_state=93, 
                                                                        test_size=0.1, stratify=target_labels)

X_train_1, X_val_1, y_train_1, y_val_1 = train_test_split(X_resid_train_1, y_resid_train_1, random_state=93,
                                                          test_size=0.2, stratify=y_resid_train_1)
```

```python
print(X_train_1.shape)
print(y_train_1.shape)
print(all_lm[0].shape[0], all_lm[0].shape[1])
print(X_val_1.shape)
```

```python
X_val_1.shape
```

```python
log_dir = os.path.join('Logs')
tb_callback = TensorBoard(log_dir=log_dir)
```

When building our model, we want to train lots of epochs, and keep track of the best performing model. The below cell instantiates a callback that saves the model's weights every time the validation loss is improved. 

```python
# Set path to folder where models will be saved
ROOT_FOLDER_1 = "./model1"
if not os.path.isdir(ROOT_FOLDER_1):
    os.makedirs(ROOT_FOLDER_1)

# create file name for each subsequent model
filepath = os.path.join(ROOT_FOLDER_1, "clf_model1-{epoch:02d}-{val_loss:.2f}.hdf5")

# create checkpoint callback to be passed to LSTM
checkpoint = tf.keras.callbacks.ModelCheckpoint(filepath, monitor='val_loss', verbose=1, save_best_only=True, mode='min')
```

```python
# random seeds for reproducibility
np.random.seed(93)
tf.keras.utils.set_random_seed(93)

callback = tf.keras.callbacks.EarlyStopping(monitor='val_loss', mode='min', patience=40)

# Instantiate Sequential neural network
clf_model_1 = Sequential()
# Add first LSTM layer with sample input shape - each video observation has 40 frames with 1662 landmark values
clf_model_1.add(LSTM(64, return_sequences=True, activation='relu', input_shape=(all_lm[0].shape[0], all_lm[0].shape[1])))
# Add second LSTM layer
clf_model_1.add(LSTM(128, return_sequences=True, activation='relu'))
# Add third LSTM layer
clf_model_1.add(LSTM(64, return_sequences=False, activation='relu'))
# Add fourth layer (Dense)
clf_model_1.add(Dense(64, activation='relu'))
# Add fifth dense layer
clf_model_1.add(Dense(32, activation='relu'))
# Add final dense layer with softmax activation function to provide predictions
clf_model_1.add(Dense(6, activation='softmax'))
```

```python
optimizer = keras.optimizers.Adam(learning_rate=0.0001)
clf_model_1.compile(optimizer=optimizer, loss="categorical_crossentropy", metrics=['categorical_accuracy'])
history1 = clf_model_1.fit(X_train_1, y_train_1, epochs=1000, callbacks=[checkpoint], validation_data=(X_val_1, y_val_1))
```

Looking at the saved models, we see that the best validation loss was created on epoch 398 with a 0.21 categorical cross entropy loss. This model achieved a train test accuracy of 98.9% and a validation accuracy of ~95.6%. The model is likely overfitting as we'll see later on that the test accuracy drops to ~92%. Regardless, it's a pretty good start. 

```python
print('The best accuracy achieved is ', round(history1.history['categorical_accuracy'][397]*100, 2), "%")
print('The best accuracy achieved is ', round(history1.history['val_categorical_accuracy'][397]*100, 2), "%")
```

Below is the graph of the model's performance over the 1000 epoch training phase, with the test and validation loss on the top graph, and the test and validation accuracies on the bottom graph. 

We see some **VERY** weird behavior with the model undergoing three very large performance drops. The last performance decline lands the model in a very bad local minimum where the loss & accuracies never recover. I'm not sure why this is happening, but it could have something to do with the learning rate, overfitting, or the individual weights being very sensitive to change.

```python
plt.figure(figsize=(9,7))
plt.plot(history1.history['loss'], label='Train Loss')
plt.plot(history1.history['val_loss'], label='Val Loss')
plt.plot([398], [0.21], marker="*", markersize=8, color='red', label='best model')
plt.xlabel('Epoch Number')
plt.ylabel('Categorical Cross Entropy Loss')
plt.title('Model Loss Vs. Training Epoch')
plt.legend()
sns.despine()
plt.show()


plt.figure(figsize=(9,7))
plt.plot(history1.history['categorical_accuracy'], label='Train Loss')
plt.plot(history1.history['val_categorical_accuracy'], label='Val Loss')
plt.plot([398], history1.history['val_categorical_accuracy'][397], label='Best Model', 
         color='red', marker="*", markersize=8)
plt.xlabel('Epoch Number')
plt.ylabel('Categorical Accuracy')
plt.title('Model Accuracy Vs. Training Epoch')
plt.legend()
sns.despine()
plt.show()
```

As the training of the model gets very weird after epoch 450, and the best model occured at 398, let's only look at the model's training performance up until epoch 400.

```python
plt.figure(figsize=(9,7))
plt.plot(history1.history['loss'][:400], label='Train Loss')
plt.plot(history1.history['val_loss'][:400], label='Val Loss')
plt.plot([398], [0.21], marker="*", markersize=8, color='red', label='best model')
plt.xlabel('Epoch Number')
plt.ylabel('Categorical Cross Entropy Loss')
plt.title('Model Loss Vs. Training Epoch')
plt.legend()
sns.despine()
plt.show()


plt.figure(figsize=(9,7))
plt.plot(history1.history['categorical_accuracy'][:400], label='Train Loss')
plt.plot(history1.history['val_categorical_accuracy'][:400], label='Val Loss')
plt.plot([398], history1.history['val_categorical_accuracy'][397], label='Best Model', 
         color='red', marker="*", markersize=8)
plt.xlabel('Epoch Number')
plt.ylabel('Categorical Accuracy')
plt.title('Model Accuracy Vs. Training Epoch')
plt.legend()
sns.despine()
plt.show()
```

The training and validation accuracies have a very tight spread up until around epoch ~150. The accuracy gap between the two widens a bit after that, suggesting that the model is overfitting. This can be solved by adding some drop out layers and regularization.

Still unsure of what is causing the random sharp declines in performance, but let's load in the best performing model and evaluate its performance on the test set.

```python
# load in best performing model
best_model_path1 = os.path.join(ROOT_FOLDER_1, 'clf_model1-398-0.21.hdf5') 
first_model = keras.models.load_model(best_model_path1)
```

Looking at the above test statistics, we see that the high overall accuracy score shows that the model is actually doing a pretty decent job generalizing on unseen data, despite some possible overfitting. The model has good precision and recall with both scores around ~93%. This means that the model is not only able to pick out each particular action being performed, but its predictions are also very **accurate**.

When we break down the performance on a per class basis, we see that all but one class has a precison above 90%. This action is the "finger" gesture, and has a precision of 77.7% and recall of 81%. This means that the model is:

1) Predicting the "middle finger" for when other actions are being performed. In particular, we see that the model predicts the middle finger action frequently for the wave and kiss action; and,
2) When the "middle finger" action is being performed, other actions are being predicted at a higher rate. Again, the wave and kiss classes are being mixed up here. 

Apart from the troublesome "finger" gesture, we see that all other classes are being correctly separated from each other. We will attempt to fix this issue with some new feature engineering later on.

## LSTM Neural Network 2 - Landmark Location Data, Excluding Face Landmarks

As our EDA has shown, the face landmarks have the lowest amount of variation in their data. Therefore, they should not be strong predictors of our target variable. As the most amount of data in our dataset comes from the face landmarks, we are going to remove the vast majority of these and only keep the face landmarks generated by the pose groupings. This will significantly reduce our model complexity, without losing a lot of predictive power (hopefully).

```python
# only take left hand, right hand, and pose landmarks
no_face_lm = pd.concat([left_copy, rh_copy, pose_copy], axis=1)
print(no_face_lm.shape)

# convert back to 3d array
no_face_lm = mat_to_3d(no_face_lm)

# sanity check
print(no_face_lm.shape)
```

```python
# train test split - target labels remain the same. Only thing we've done is removed features so only need to update
# X dataframe (no_face_lm in this case)
X_resid_train_2, X_test_2, y_resid_2, y_test_2 = train_test_split(no_face_lm, target_labels, random_state=93, 
                                                                  stratify = target_labels, test_size=0.1)

X_train_2, X_val_2, y_train_2, y_val_2 = train_test_split(X_resid_train_2, y_resid_2, random_state=93,
                                                          stratify = y_resid_2, test_size=0.2)
```

```python
# Set path to folder where models will be saved
ROOT_FOLDER_2 = "./model2"
if not os.path.isdir(ROOT_FOLDER_2):
    os.makedirs(ROOT_FOLDER_2)

# create file name for each subsequent model
filepath = os.path.join(ROOT_FOLDER_2, "clf_model2-{epoch:02d}-{val_loss:.2f}.hdf5")

# create checkpoint callback to be passed to LSTM
checkpoint = tf.keras.callbacks.ModelCheckpoint(filepath, monitor='val_loss', verbose=1, save_best_only=True, mode='min')
```

```python
# random seeds for reproducibility
np.random.seed(93)
tf.keras.utils.set_random_seed(93)

# set early stoppage
callback = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=30, mode='min')

# Instantiate Sequential neural network
clf_model_2 = Sequential()
# Add first LSTM layer with sample input shape - each video observation has 40 frames with 1662 landmark values
clf_model_2.add(LSTM(64, return_sequences=True, activation='relu', 
                     input_shape=(no_face_lm[0].shape[0], no_face_lm[0].shape[1])))
# Add second LSTM layer
clf_model_2.add(LSTM(128, return_sequences=True, activation='relu'))
# Add third LSTM layer
clf_model_2.add(LSTM(64, return_sequences=False, activation='relu'))
# Add fourth layer (Dense)
clf_model_2.add(Dense(64, activation='relu'))
# Add fifth dense layer
clf_model_2.add(Dense(32, activation='relu'))
# Add final dense layer with softmax activation function to provide predictions
clf_model_2.add(Dense(6, activation='softmax'))
```

When we graph the model performance vs epoch below, we see that the best validation accuracy and smallest validation accuracy happen at two different places... 

Looking at the graphs we see that the model converges on a local minimum, around epoch 200, as the loss and accuracies for both train and validation sets don't change for nearly 200 epochs. ## LSTM Neural Network 3 - Hand Distance Data

For the next model I've engineered some new distance features in order to better separate the finger action from the others. When the distance between the tip of the middle finger and wrist is graphed over time, we observe a unique pattern which separates the middle finger from the other actions. I'm hoping that these additional distance features will help model performance, in particular with distinguishing the finger action.

I will first train a model just on the distance data before combining it with the other landmark data.

```python
# create distance copy
X_dist = lm_distances.drop(['action', 'video sample number', 'frame number'], axis=1).copy()

# convert back to 3d array
X_dist = mat_to_3d(X_dist)

# sanity check
print(X_dist.shape)
```

```python
# Set path to folder where models will be saved
ROOT_FOLDER_3 = "./model3"
if not os.path.isdir(ROOT_FOLDER_3):
    os.makedirs(ROOT_FOLDER_3)

# create file name for each subsequent model
filepath = os.path.join(ROOT_FOLDER_3, "clf_model3-{epoch:02d}-{val_categorical_accuracy:.2f}.hdf5")

# create checkpoint callback to be passed to LSTM
checkpoint = tf.keras.callbacks.ModelCheckpoint(filepath, monitor='val_categorical_accuracy', verbose=1,
                                                save_best_only=True, mode='max')
```

```python
# train test split - target labels remain the same. Only thing we've done is removed features so only need to update
# X dataframe (no_face_lm in this case)

X_resid_train_4, X_test_4, y_resid_train_4, y_test_4 = train_test_split(X_dist, target_labels, random_state=93, 
                                                                        test_size=0.1)

X_train_4, X_val_4, y_train_4, y_val_4 = train_test_split(X_resid_train_4, y_resid_train_4, random_state=93, 
                                                                        test_size=0.2)
```

```python
# random seeds for reproducibility
np.random.seed(93)
tf.keras.utils.set_random_seed(93)


# Instantiate Sequential neural network
clf_model_4 = Sequential()
# Add first LSTM layer with sample input shape - each video observation has 40 frames with 1662 landmark values
clf_model_4.add(LSTM(64, return_sequences=True, activation='relu', 
                     input_shape=(X_dist[0].shape[0], X_dist[0].shape[1])))
# Add second LSTM layer
clf_model_4.add(LSTM(128, return_sequences=True, activation='relu'))
# Add third LSTM layer
clf_model_4.add(LSTM(64, return_sequences=False, activation='relu'))
# Add fourth layer (Dense)
clf_model_4.add(Dense(64, activation='relu'))
# Add fifth dense layer
clf_model_4.add(Dense(32, activation='relu'))
# Add final dense layer with softmax activation function to provide predictions
clf_model_4.add(Dense(6, activation='softmax'))
```

```python
optimizer = keras.optimizers.Adam(learning_rate=0.0001)
clf_model_4.compile(optimizer=optimizer, loss="categorical_crossentropy", metrics=['categorical_accuracy'])
clf_model_4.fit(X_train_4, y_train_4, epochs=1000, callbacks=[checkpoint], validation_data = (X_val_4, y_val_4))
```

```python
yhat4 = clf_model_4.predict(X_test_4)
ytrue4 = np.argmax(y_test_4, axis=1).tolist()
yhat4 = np.argmax(yhat4, axis=1).tolist()
ytruelabels4 = [get_label_string(x) for x in yhat4]

cf_matrix = confusion_matrix(ytrue4, yhat4)
print('accuracy score of {}'.format(accuracy_score(ytrue4, yhat4)))
```
